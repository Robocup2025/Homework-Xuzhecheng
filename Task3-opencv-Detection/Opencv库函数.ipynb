```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread("messi5.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


low_threshold = 50
high_threshold = 100
edges = cv2.Canny(gray, low_threshold, high_threshold)

# 
plt.figure(figsize=(20,10))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)); plt.title("Original"); plt.axis("off")
plt.subplot(1,2,2); plt.imshow(edges, cmap='gray'); plt.title("CannyDetection"); plt.axis("off")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread("messi5.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

gray = np.float32(gray)
dst = cv2.cornerHarris(gray, blockSize=2, ksize=3, k=0.04)
dst = cv2.dilate(dst, None)

corner_img = img.copy()
corner_img[dst > 0.01 * dst.max()] = [0, 255, 0]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(corner_img, cv2.COLOR_BGR2RGB))
plt.title("Harris Corners (OpenCV)")
plt.axis("off")

plt.tight_layout()
plt.show()
```
